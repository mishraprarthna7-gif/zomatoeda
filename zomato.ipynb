```python
df_country=pd.read_excel('C:\\data sc\\Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(4)
```

```python
final_df.dtypes # to check data types
```

```python
country_names=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts().values
#pie chart top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

observation: zomato max records or transaction are from india after usa and then uk

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show()
```

```python

sns.barplot (x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
plt.show()
```

observation :
not rated count is very high
max number of rating are btw 2.5 to 3.4


```python
# count plot
sns.countplot(x="Rating color",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
plt.show()
```
